```
import tokenizers
```

```
bwpt = tokenizers.BertWordPieceTokenizer(
    vocab_file=None,
    add_special_tokens=True,
    unk_token='[UNK]',
    sep_token='[SEP]',
    cls_token='[CLS]',
    clean_text=True,
    handle_chinese_chars=False,
    strip_accents=True,
    lowercase=True,
    wordpieces_prefix='##'
)
```

```
bwpt.train(
    files=[file_path],
    # files = ["bert_thesis_experiments/working/mysample.txt"],
    vocab_size=30522,
    min_frequency=2,
    limit_alphabet=1000,
    special_tokens=['[PAD]', '[UNK]', '[CLS]', '[MASK]', '[SEP]']
)
```

```
bwpt.save("bert_thesis_experiments/working", "protein_seq_words_uniref0.5_vsz30k_mfq2")
```
